```python
#Input elements of 2 matrices
A = [[1,2],[2,3]]
B = [[1,2,3,4,5],[5,6,7,8,9]]

row_A = len(A)
col_A = len(A[0])
row_B = len(B)
col_B = len(B[0]) 

#Initialize a null matrix 
E = [[0 for col in range(col_B)] for row in range(row_A)]

def multi_matr(A,B):
    '''Multiplying 2 matrices'''
    if col_A == row_B:
        for i in range(row_A):
            for j in range(col_B):
                for k in range(row_B):
                    E[i][j] += A[i][k]*B[k][j]
    return E

if (len(A[0]) == len(B)):
    print("Result of Multiplication is: {}".format(multi_matr(A,B)))
else:
    print("Multiplication is not possible")



        

    
```

```python
import math
S = [(1,2),(3,4),(-1,1),(6,-7),(0,6),(-5,-8),(-1,-1),(6,0),(1,-1)]   
P = (3,-4)
def find_cos(x):
    return math.acos((x[0]*P[0]+x[1]*P[1])/((math.sqrt(x[0]**2+x[1]**2))*(math.sqrt(P[0]**2+P[1]**2))))
D = list(map(find_cos,S))
E = {D[i]:S[i] for i in range(len(S))}
A = list(E.keys())
A.sort()
for i in range(5):
    print(E[A[i]])
   
```

```python
import re
Red = [(1,1),(2,1),(4,2),(2,4),(-1,4)]
Blue = [(-2,-1),(-1,-2),(-3,-2),(-3,-1),(1,-3)]
lines = ["1x+1y+0","1x-1y+0","1x+0y-3","0x+1y-0.5"]
K=[];T=[];R=[];L=[];D=[]
for line in lines:
    K.append(re.split('[xy]+',line))
for i in K:
    for j in Red:
        D.append(j[0]*float(i[0])+j[1]*float(i[1])+float(i[2]))
    R.append(D)
    D=[]
for i in K:
    for j in Blue:
         T.append(j[0]*float(i[0])+j[1]*float(i[1])+float(i[2]))
    L.append(T)
    T=[]
for i in range(len(L)):
    D.append(all(x > 0 for x in R[i]))
    T.append(all(x > 0 for x in L[i]))

for i in range(len(D)):
    if D[i]== True and T[i]== False:
        print("Yes")
    elif D[i]== False and T[i]== True:
        print("Yes")
    else:
        print("No")
            
        
        


    
    
    
    
```
